```python
titanic.isnull().sum()
```

```python
titanic['Age'].fillna(titanic['Age'].mean(),inplace = True) #fill age with mean
```

```python
titanic['Embarked'].fillna(titanic['Embarked'].mode(0),inplace = True) # fill missing value in embarked with mode
```

```python
titanic.drop('Cabin',axis=1,inplace = True) #to many nulls
```

```python
import numpy as np
```

```python
#create age group columns
titanic['Age_group'] = np.where(titanic['Age']<18, 'Child', np.where(titanic['Age']<60, 'Adult', 'Senior'))
```

```python
titanic['Age_group'].value_counts()
```

```python
#convert survived from 0/1 to text
titanic['SurvivalStatus'] = titanic['Survived'].map({0:'Died',1:'Survived'})
```

```python
titanic['SurvivalStatus'].value_counts()
```

```python
#female passengers who survived
print(titanic[(titanic['Sex']=='female') & (titanic['Survived'] ==1)])
```

```python
print(titanic[(titanic['Age']>50) & (titanic['Pclass'] ==1)]) #passenger older than 50 and in first class
```

```python
#survival rate by class
print(titanic.groupby('Pclass')['Survived'].mean())
```

```python
#print name of passengers under 10 years of age
for name in titanic[titanic['Age']<10]['Name']:
    print(name)
```

```python
#save cleaned file
titanic_cleaned = titanic
```

```python
titanic_cleaned.to_csv('titanic_cleaned_data.csv', index=False) 
```
